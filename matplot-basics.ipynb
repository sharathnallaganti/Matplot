```python
import matplotlib.pyplot as plt


x = [2,4,6,8,10,12]

y = [1,3,6,9,12,15]

"""
Types of graphs 
plt.plot()  line plot

plt.scatter()  scatter plot

plt.bar()  bar chart

plt.hist()  histogram

plt.subplots()  multiple plots

"""

plt.plot(x,y)
plt.title("Weather Forecast")
plt.xlabel("Temperature")
plt.ylabel("Humidity")
plt.show()

# Scatter Plot
plt.scatter(x,y)
plt.show()

# Bar Chart
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 4]

plt.bar(categories, values)
plt.title("Bar Chart")
plt.show()


# 
```
