```python
import matplotlib.pyplot as plt
import numpy as np


x = [2,4,6,8,10,12]

y = [1,3,6,9,12,15]

"""
Types of graphs 
plt.plot() → line plot

plt.scatter() → scatter plot

plt.bar() → bar chart

plt.hist() → histogram

plt.subplots() → multiple plots

"""

plt.plot(x,y)
plt.title("Weather Forecast")
plt.xlabel("Temperature")
plt.ylabel("Humidity")
plt.show()

# Scatter Plot
plt.scatter(x,y)
plt.show()

# Bar Chart
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 4]

plt.bar(categories, values)
plt.title("Bar Chart")
plt.show()

# Histograms

data = np.random.rand(1000)
plt.hist(data, bins=40 ,color='skyblue', edgecolor='black', alpha=0.7)
plt.title("Histogram")
plt.show()


# Pi chart

ethinicites = ["Asians" , "latin", "white", "black", "hispanic"]

percentages = [20, 15, 25, 30, 10]

plt.pie (percentages, 
         labels = ethinicites,
         autopct = "%1.1f%%",
         startangle=90,
         colors=['#66b3ff','#99ff99','#ffcc99','#ff9999','#c2c2f0'], 
        shadow=True
        
        )

plt.title("Ethnicity Distribution")
plt.show()
```
